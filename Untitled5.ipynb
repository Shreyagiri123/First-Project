```python
import random as r

ProductList = {
    'p1':10, 'p2':15, 'p3':20, 'p4':25, 'p5':30, 'p6':35, 'p7':50,
    'p8':40, 'p9':55, 'p10':60, 'p11':65, 'p12':75, 'p13':70, 'p14':45
}

LB = 90
UB = 210
ResultList = set()
Iterations = 1000

for _ in range(Iterations):
    SetSize = r.randint(2, len(ProductList))  # Range is now 2 to total keys
    ComboList = r.sample(list(ProductList.keys()), SetSize)
    ComboList.sort()
    ComboSum = sum([ProductList[i] for i in ComboList])
    if LB <= ComboSum <= UB:
        ResultList.add(tuple(ComboList))

for item in ResultList:
    print(item)

print("\nTotal Sets:", len(ResultList), "\n")
```
