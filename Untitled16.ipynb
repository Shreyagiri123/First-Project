```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor, plot_importance
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('/content/train (2).csv')
test = pd.read_csv('/content/test (2).csv')
```

```python
X = train.drop(['id', 'yield'], axis=1)
y = train['yield']
X_test = test.drop('id', axis=1)
```

```python
Q1 = y.quantile(0.25)
Q3 = y.quantile(0.75)
IQR = Q3 - Q1
mask = (y >= Q1 - 1.5 * IQR) & (y <= Q3 + 1.5 * IQR)
X = X[mask]
y = y[mask]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5],
    'learning_rate': [0.05, 0.1]
}
```

```python
model = XGBRegressor(random_state=42)
grid = GridSearchCV(model, param_grid, scoring='neg_mean_absolute_error', cv=3)
grid.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid.best_params_)
print("Best Validation MAE:", -grid.best_score_)
```

```python
best_model = grid.best_estimator_
best_model.fit(X_scaled, y)
```

```python
test_preds = best_model.predict(X_test_scaled)
```

```python
plt.figure(figsize=(10, 6))
plot_importance(best_model, max_num_features=10)
plt.title("Top 10 Important Features")
plt.show()
```

```python
submission = pd.DataFrame({
    'id': test['id'],
    'yield': test_preds
})
```

```python
submission.to_csv('Shreya_Giri_102303684.csv', index=False)
print("✅ Submission file created successfully.")
```
